```python
berat = float(input("Berapakah berat beras?"))
kg_atau_ton = input("Berat beras dalam kg atau ton?")

if kg_atau_ton.upper() == 'KG':
    berat = berat*1000
    print("Berat beras adalah"+str(berat) + "ton")
elif kg_atau_ton.upper() == 'TON':
    berat = berat/1000
    print(f"Berat beras adalah {berat} kg")
```

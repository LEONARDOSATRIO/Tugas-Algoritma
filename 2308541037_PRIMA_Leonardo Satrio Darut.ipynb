```python
# Mencari bilangan prima jika diketahui n= 37 (dua digit terakhir nim saya)
# Maka kita harus menentukan bilangan prima antara 1000n dan 1000(n+1) atau 37000 dan 38000
def is_prime(n_37):
    if n_37 <= 1:
        return False
    if n_37 <= 3:
        return True
    if n_37 % 2 == 0 or n_37 % 3 == 0:
        return False
    i_37 = 5
    while i_37 * i_37 <= n_37:
        if n_37 % i_37 == 0 or n_37 % (i_37 + 2) == 0:
            return False
        i_37 += 6
    return True

def find_primes_in_range(start, end):
    prime_numbers = []
    for num in range(start, end + 1):
        if is_prime(num):
            prime_numbers.append(num)
    return prime_numbers


# Contoh penggunaan:
start = 37000
end = 38000
primes = find_primes_in_range(start, end)
print("Bilangan prima antara", start, "dan", end, "adalah:", primes)
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Persamaan garis pertama melalui dua titik yaitu (a,b) dan (c,d)
a_37 = float(input('=> Nilai x dari titik pertama garis pertama adalah'))
b_37 = float(input('=> Nilai y dari titik pertama garis pertama adalah'))
c_37 = float(input('=> Nilai x dari titik kedua garis pertama adalah'))
d_37 = float(input('=> Nilai y dari titik kedua garis pertama adalah'))
# Gradien garis pertama adalah
m1_37 = (d_37-b_37)/(c_37-a_37)
# Menentukan titik potong garis pertama
c1_37 = -m1_37*a_37+b_37
print('=> Persamaan garis pertama adalah y={}x+{}'.format(m1_37, c1_37))


# Persamaan garis kedua melalui dua titik yaitu (e,f) dan (g,h)
e_37 = float(input('=> Nilai x dari titik pertama garis kedua adalah'))
f_37 = float(input('=> Nilai y dari titik pertama garis kedua adalah'))
g_37 = float(input('=> Nilai x dari titik kedua garis kedua adalah'))
h_37 = float(input('=> Nilai y dari titik kedua garis kedua adalah'))
# Gradien garis kedua adalah
m2_37 = (h_37-f_37)/(g_37-e_37)
# Menentukan titik potong garis kedua
c2_37 = -m2_37*e_37+f_37
print('=> Persamaan garis kedua adalah y={}x+{}'.format(m2_37, c2_37))

if m1_37==m2_37:
    # Terdapat kemungkinan garisnya berhimpit atau sejajar
    if c1_37==c2_37:
        print('=> Kedua garis berhimpitan dan memiliki tak hingga titik potong sehingga meiliki tak hingga solusi')
        x_37=np.linspace(-4,6,100)
        plt.xlabel("x") 
        plt.ylabel("y")
        plt.plot(x_37,m1_37*(x_37-a_37)+b_37,color="red",label="y={}x+{}".format(m1_37,-m1_37*a_37+b_37))
        plt.plot(x_37,m2_37*(x_37-e_37)+f_37,color="blue",label="y={}x+{}".format(m2_37,-m2_37*e_37+f_37))
        plt.legend(loc="best")
        plt.grid(True)
        plt.show()
    else:
        print('=> kedua garis sejajar dan  tidak memiliki titik potong, sehingga tidak memiliki solusi')
        x_37=np.linspace(-4,6,100)
        plt.xlabel("x") 
        plt.ylabel("y")
        plt.plot(x_37,m1_37*(x_37-a_37)+b_37,color="red",label="y={}x+{}".format(m1_37,-m1_37*a_37+b_37))
        plt.plot(x_37,m2_37*(x_37-e_37)+f_37,color="green",label="y={}x+{}".format(m2_37,-m2_37*e_37+f_37))
        plt.legend(loc="best")
        plt.grid(True)
        plt.show()
        
else:
    print('=> terdapat satu titik potong sehingga solusinya hanya satu')
    tp_x_37 = (m1_37*a_37-b_37-m2_37*e_37+f_37)/(m1_37-m2_37) #titik potong persamaan garis 1
    tp_y_37 = m2_37*(tp_x_37-e_37)+f_37 #titik potong persamaan garis 2
    print ("jadi titik potong kedua garis adalah ({},{})".format(tp_x_37,tp_y_37))

    x_37=np.linspace(tp_x_37 - 6, tp_x_37 + 6, 200)
    plt.title("Gambar Titik Potong Persamaan Garis")
    plt.xlabel("x") 
    plt.ylabel("y")
    plt.plot(x_37,m1_37*(x_37-a_37)+b_37,color="red",label="y1 = {}x+{}".format(m1_37,-m1_37*a_37+b_37))
    plt.plot(x_37,m2_37*(x_37-e_37)+f_37,color="green",label="y2 = {}x+{}".format(m2_37,-m2_37*e_37+f_37))
    plt.plot(tp_x_37,tp_y_37,color="black",marker="o",markersize=5) 
    plt.legend(loc="best")
    plt.grid(True)
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definisikan dua pasang titik
# Garis pertama
x1_37 = 2
y1_37 = 3
x2_37 = 5
y2_37 = 7
# Garis kedua
x3_37 = 1
y3_37 = 5
x4_37 = 8
y4_37 = 4

# Fungsi untuk menghitung koefisien a, b, dan c dari persamaan garis ax + by = c
def compute_line_through_points(x1_37, y1_37, x2_37, y2_37):
    a_37 = y2_37 - y1_37
    b_37 = x1_37 - x2_37
    c_37 = a_37 * x1_37 + b_37 * y1_37    
    return a_37, b_37, c_37

# Hitung koefisien untuk kedua garis
a1_37, b1_37, c1_37 = compute_line_through_points(x1_37, y1_37, x2_37, y2_37)
a2_37, b2_37, c2_37 = compute_line_through_points(x3_37, y3_37, x4_37, y4_37)

# Print persamaan garis dalam bentuk ax + by = c
print(f"Persamaan garis pertama melalui titik ({x1_37}, {y1_37}) dan ({x2_37}, {y2_37}) adalah: {a1_37}x + {b1_37}y = {c1_37}")
print(f"Persamaan garis kedua melalui titik ({x3_37}, {y3_37}) dan ({x4_37}, {y4_37}) adalah: {a2_37}x + {b2_37}y = {c2_37}")

# Fungsi untuk menggambar garis
def plot_line(a_37, b_37, c_37, x_range, label):
    x_37 = np.linspace(x_range[0], x_range[1], 400)
    y_37 = (c_37 - a_37 * x_37) / b_37
    plt.plot(x_37, y_37, label=label)
    plt.xlabel('x_37')
    plt.ylabel('y_37')
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)

# Plot titik-titik
plt.scatter([x1_37, x2_37, x3_37, x4_37], [y1_37, y2_37, y3_37, y4_37], color='red', zorder=5)
plt.text(x1_37, y1_37, f'({x1_37}, {y1_37})', fontsize=12, ha='right')
plt.text(x2_37, y2_37, f'({x2_37}, {y2_37})', fontsize=12, ha='right')
plt.text(x3_37, y3_37, f'({x3_37}, {y3_37})', fontsize=12, ha='right')
plt.text(x4_37, y4_37, f'({x4_37}, {y4_37})', fontsize=12, ha='right')

# Plot garis berdasarkan persamaan
plot_line(a1_37, b1_37, c1_37, x_range=(-10, 10), label=f'{a1_37}x + {b1_37}y = {c1_37}')
plot_line(a2_37, b2_37, c2_37, x_range=(-10, 10), label=f'{a2_37}x + {b2_37}y = {c2_37}')

# Tampilkan plot
plt.legend()
plt.title('Garis Saling Berpotongan Melalui Dua Pasang Titik')
plt.show()
```
